#### Task 3b: Create Derived Columns from Datetime

Add two categorical columns to the rides DataFrame based on trip start time:

2. `time_of_day`: String categories based on start hour:
   - "Morning Rush" if hour is 7, 8, or 9
   - "Evening Rush" if hour is 16, 17, or 18
   - "Midday" for all other hours

````python
# Task 3b code here...
# make sure the index is datetime (it should be from 2d)
rides.index = pd.to_datetime(rides.index)

# 1) weekend flag: Saturday=5, Sunday=6
rides['is_weekend'] = rides.index.dayofweek.isin([5, 6])

# 2) time_of_day: start with "Midday"
rides['time_of_day'] = "Midday"

hours = rides.index.hour

# morning rush: 7, 8, 9
morning_mask = hours.isin([7, 8, 9])
rides.loc[morning_mask, 'time_of_day'] = "Morning Rush"

# evening rush: 16, 17, 18
evening_mask = hours.isin([16, 17, 18])
rides.loc[evening_mask, 'time_of_day'] = "Evening Rush"

# 3) show distribution
rides['time_of_day'].value_counts()


````

### Interpretation – Problem 3

**1. `isin()`**
If I only have 2–3 values, doing `(col == 'A') | (col == 'B')` is fine. But if I have a bunch (like 10–20 station names), `col.isin(list_of_stations)` is way better. I can make the list once and reuse it, and I’m less likely to mess up a `|` or a quote.

**3. `query()` vs boolean**
`query()` is nice when the condition is simple and I want it to read clearly, like: “casual and weekend and duration > 20.” I’d pick boolean indexing when I’m using variables or building the filter in code, because that’s easier to write in plain Python. So: `query()` = nicer to read, boolean = more flexible.

### Graduate Follow-Up 


Pandas kind of lets you do a bunch of things with `df[...]`, but that’s also why it gets confusing. The rule they gave us — use `df['col']` for columns, and use `loc`/`iloc` for rows — just makes it clearer what we meant.

- `df['col']` → I want this column.

````python
# Task 4a code here...

# make sure index is datetime
rides.index = pd.to_datetime(rides.index)

# 1) most popular hour
hour_counts = rides.index.hour.value_counts()
peak_hour = hour_counts.idxmax()
peak_hour_count = hour_counts.max()
print(f"Most popular hour: {peak_hour}:00 with {peak_hour_count} trips")

# 2) busiest day
day_counts = rides.index.day_name().value_counts()
busiest_day = day_counts.idxmax()
busiest_day_count = day_counts.max()
print(f"Busiest day: {busiest_day} with {busiest_day_count} trips")

# 3) sort Monday → Sunday using sort_index()
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

day_counts.index = pd.CategoricalIndex(day_counts.index,
                                       categories=weekday_order,
                                       ordered=True)

day_counts_sorted = day_counts.sort_index()

print("\nTrips by weekday (Mon→Sun):")
print(day_counts_sorted)


````

### Interpretation – Problem 4

**1. What `groupby()` is really doing**
I think of `groupby()` as “split → apply → combine.” First it **splits** the data into buckets (one bucket for members, one for casual). Then it **applies** whatever math I asked for to each bucket (like average trip duration). Then it **combines** those results back into one small table so I can read it. So it’s not doing magic — it’s just doing the same calc multiple times, once per group.

**2. `value_counts()` vs `groupby().size()`**
If I just want “how many members vs how many casual?” I’ll do `rides['member_casual'].value_counts()` — it’s shorter. If I’m already grouping, or I think I might group on more than one thing later, I’ll use `rides.groupby('member_casual').size()`. The second one is more flexible. Also, if I needed counts by *two* columns, `value_counts()` on one column wouldn’t be enough, but `groupby()` would work.

**3. Why we did `reset_index()` before export**
Our time was sitting in the index. When we write a CSV with `index=False`, the index gets dropped. So if we didn’t reset it, we’d lose the timestamp in the file. `reset_index()` just turns that index back into a normal column (`started_at`) so it actually shows up in the CSV.

Compare `CSV` and _pickle_ formats for data storage and retrieval.

Pickle is Python's built-in serialization format that saves Python objects exactly as they exist in memory, preserving all data types, structures, and metadata. Unlike CSV (which converts everything to text), pickle is binary (not human readable) and maintains the complete state of your DataFrame. Also, pickle files only work in Python, while CSV is universal. Read more in the [Pandas documentation](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_pickle.html).
